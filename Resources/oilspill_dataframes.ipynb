```python
#import dependencies and libraries
import pandas as pd
import numpy as np
from dataprep.clean import clean_lat_long
from dataprep.clean import validate_lat_long
```

```python
#define function to extract chosen columns
def extract(pipeline_file,column_names):
    #Read data into df
    pipeline_data = pd.read_csv(pipeline_file, low_memory=False)
    fixed_pipeline_data = pipeline_data[column_names]
    return pipeline_data, fixed_pipeline_data
```

### Pre-1986

```python
# Load pre-1986 data
column_names = ['REPORT_ID','OPERATOR_NAME','ACCIDENT_DATE','ACCIDENT_STATE', 'ACCIDENT_COUNTY', 
    'ACCIDENT_CITY', 'COMMODITY', 'LOSS_BBLS', 'TOTAL_FATALITIES', 'TOTAL_INJURIES', 
    'IS_FIRE', 'IS_EXPLOSION', 'IS_ABOVE_GROUND_TEXT','TOTAL_PRPTY_DAMAGE','ESTIMATE_ACC_PRESSURE',
    'DEPTH_OF_COVER','SYS_PART_INVOLVED_TEXT', 'ORIGIN_LIQUID_RELEASE_TEXT', 'PRIMARY_CAUSE_TEXT']

pipeline_file = 'pre1986.csv'
pipeline_data, fixed_pipeline_data = extract(pipeline_file,column_names)
```

```python
# Convert pre-1986 data to dataframe
pre1986_df = fixed_pipeline_data
pre1986_df
```

```python
#create list of old column names and new column names
old_names = [
    'REPORT_ID',
    'OPERATOR_NAME',
    'ACCIDENT_DATE',
    'ACCIDENT_STATE', 
    'ACCIDENT_COUNTY', 
    'ACCIDENT_CITY', 
    'COMMODITY', 
    'TOTAL_FATALITIES', 
    'TOTAL_INJURIES',
    'IS_ABOVE_GROUND_TEXT', 
    'DEPTH_OF_COVER',
    'SYS_PART_INVOLVED_TEXT', 
    'TOTAL_PRPTY_DAMAGE',
    'ESTIMATE_ACC_PRESSURE', 
    'PRIMARY_CAUSE_TEXT']

new_names = [
 'REPORT_NUMBER',
 'NAME',
 'DATE_OF_INCIDENT',
 'STATE OF INCIDENT',
 'COUNTY OF INCIDENT',
 'CITY OF INCIDENT',
 'COMMODITY_RELEASED_TYPE',
 'FATALITY_IND',
 'INJURY_IND',
 'INCIDENT_AREA_TYPE',
 'DEPTH_OF_COVER',
 'ITEM_INVOLVED',
 'PRPTY',
 'ACCIDENT_PRESSURE',
 'CAUSE'
 ]
#create dictionary of old name and new names
new_columns = dict(zip(old_names, new_names))

#rename columns
pre1986_df.rename(new_columns, axis=1, inplace=True)
pre1986_df.head()   
```

```python
# convert 'DATETIME_OF_INCIDENT' column to datetime format
yr2002_2009_df['DATETIME_OF_INCIDENT'] = yr2002_2009_df['DATETIME_OF_INCIDENT'].astype('datetime64[ns]')
yr2002_2009_df
```

# Create df of total incidents per year

```python
# Create list of years
pre86 = pd.DatetimeIndex(pre1986_df['DATE_OF_INCIDENT']).year
pre86to02 = pre86.append(pd.DatetimeIndex(yr1986_2002_df['DATETIME_OF_INCIDENT']).year)
pre86to09 = pre86to02.append(pd.DatetimeIndex(yr2002_2009_df['DATETIME_OF_INCIDENT']).year)
pre86topresent = pre86to09.append(pd.DatetimeIndex(yr2010_present_df['DATE_OF_INCIDENT']).year)

# Add column to dataframe
pre86topresent_df = pd.DataFrame(pre86topresent)
pre86topresent_df.columns=['Year']
```

```python
# Group by year and add count per year
incidents_per_year = pre86topresent_df.groupby(pre86topresent_df['Year']).size()
incidents_per_year_df = pd.DataFrame(incidents_per_year)
incidents_per_year_df.columns=['Incidents_per_Year']
incidents_per_year_df.head()
```

```python
# Add bbls lost to df
released_df = pd.concat([pre1986_df['LOSS_BBLS'], yr1986_2002_df['TOTAL_BBLS'], yr2002_2009_df['TOTAL_BBLS_RELEASED'], yr2010_present_df['TOTAL_BBLS_RELEASED']], axis=0, ignore_index=True)
released_df
pre86topresent_df['TOTAL_BBLS_RELEASED'] = released_df
released_per_year = pre86topresent_df.groupby(pre86topresent_df['Year']).sum('TOTAL_BBLS_RELEASED')
released_per_year
incidents_per_year_df['TOTAL_BBLS_RELEASED'] = released_per_year['TOTAL_BBLS_RELEASED']
incidents_per_year_df.head()
```

```python
# Add bbls lost to df
costs_df = pd.concat([pre1986_df['PRPTY'], yr1986_2002_df['PRPTY'], yr2002_2009_df['TOTAL_EST_COST'], yr2010_present_df['TOTAL_EST_COST']], axis=0, ignore_index=True)
pre86topresent_df['TOTAL_COST'] = costs_df
cost_per_year = pre86topresent_df.groupby(pre86topresent_df['Year']).sum('TOTAL_COST')
incidents_per_year_df['TOTAL_COST'] = cost_per_year['TOTAL_COST']
incidents_per_year_df.head()
```

```python
# Create df with cost of each incident and associated year
cost_per_incident_df = pre86topresent_df.drop('TOTAL_BBLS_RELEASED', axis=1)
cost_per_incident_df
cost_per_incident_df.to_csv('cost_per_incident.csv', index=True)
```

```python
# Export incidents per year to csv file
incidents_per_year_df.to_csv('incidents_per_year.csv', index=True)
```

```python
# Export incidents per year to json file
incidents_per_year_df.to_json('Timeline/incidents_per_year.json')
```
