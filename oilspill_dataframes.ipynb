```python
#import dependencies and libraries
import pandas as pd
import numpy as np
```

```python
#define function to extract chosen columns
def extract(pipeline_file,column_names):
    #Read data into df
    pipeline_data = pd.read_csv(pipeline_file, low_memory=False)
    fixed_pipeline_data = pipeline_data[column_names]
    return pipeline_data, fixed_pipeline_data
```

### Pre-1986

```python

column_names = ['REPORT_ID','OPERATOR_NAME','ACCIDENT_DATE','ACCIDENT_STATE', 'ACCIDENT_COUNTY', 
    'ACCIDENT_CITY', 'COMMODITY', 'LOSS_BBLS', 'TOTAL_FATALITIES', 'TOTAL_INJURIES', 
    'IS_FIRE', 'IS_EXPLOSION', 'IS_ABOVE_GROUND_TEXT','TOTAL_PRPTY_DAMAGE','ESTIMATE_ACC_PRESSURE',
    'DEPTH_OF_COVER','SYS_PART_INVOLVED_TEXT', 'ORIGIN_LIQUID_RELEASE_TEXT', 'PRIMARY_CAUSE_TEXT']

pipeline_file = 'pre1986.csv'
pipeline_data, fixed_pipeline_data = extract(pipeline_file,column_names)
```

```python
pre1986_df = fixed_pipeline_data
pre1986_df
```

```python
#create list of old column names and new column names
old_names = [
    'REPORT_ID',
    'OPERATOR_NAME',
    'ACCIDENT_DATE',
    'ACCIDENT_STATE', 
    'ACCIDENT_COUNTY', 
    'ACCIDENT_CITY', 
    'COMMODITY', 
    'TOTAL_FATALITIES', 
    'TOTAL_INJURIES',
    'IS_ABOVE_GROUND_TEXT', 
    'DEPTH_OF_COVER',
    'SYS_PART_INVOLVED_TEXT', 
    'TOTAL_PRPTY_DAMAGE',
    'ESTIMATE_ACC_PRESSURE', 
    'PRIMARY_CAUSE_TEXT']

new_names = [
 'REPORT_NUMBER',
 'NAME',
 'DATE_OF_INCIDENT',
 'STATE OF INCIDENT',
 'COUNTY OF INCIDENT',
 'CITY OF INCIDENT',
 'COMMODITY_RELEASED_TYPE',
 'FATALITY_IND',
 'INJURY_IND',
 'INCIDENT_AREA_TYPE',
 'DEPTH_OF_COVER',
 'ITEM_INVOLVED',
 'PRPTY',
 'ACCIDENT_PRESSURE',
 'CAUSE'
 ]
#create dictionary of old name and new names
new_columns = dict(zip(old_names, new_names))

#rename columns
pre1986_df.rename(new_columns, axis=1, inplace=True)
pre1986_df.head()   
```

### 2010 - present

```python
column_names = 'REPORT_NUMBER NAME LOCAL_DATETIME LOCATION_LATITUDE LOCATION_LONGITUDE COMMODITY_RELEASED_TYPE TOTAL_BBLS_RELEASED FATALITY_IND INJURY_IND	SHUTDOWN_DUE_ACCIDENT_IND	IGNITE_IND	EXPLODE_IND	ON_OFF_SHORE	INCIDENT_AREA_TYPE	DEPTH_OF_COVER	ITEM_INVOLVED	PIPE_DIAMETER INSTALLATION_YEAR RELEASE_TYPE	LEAK_TYPE	PRPTY	ACCIDENT_PRESSURE	SCADA_DETECTION_IND	CPM_DETECTION_IND	WATER_CONTAM_IND	CAUSE	CAUSE_DETAILS'
#create list of column names
column_names = column_names.split()
```

```python

pipeline_file = 'yr2010_present.csv'
pipeline_data, fixed_pipeline_data = extract(pipeline_file,column_names)
yr2010_present_df = fixed_pipeline_data
```

```python
yr2010_present_df.columns
```

```python
new_columns= dict(LOCAL_DATETIME= 'DATE_OF_INCIDENT', PRPTY = 'TOTAL_EST_COST')
yr2010_present_df.rename(new_columns, axis=1, inplace=True)
yr2010_present_df.head()
```

```python
#convert 'DATE_OF_INCIDENT' column to datetime format

yr2010_present_df['DATE_OF_INCIDENT'] = yr2010_present_df['DATE_OF_INCIDENT'].astype('datetime64[ns]')

# Check the format of 'DATE_OF_INCIDENT' column
yr2010_present_df.info()

```

```python
#Get date ONLY from DATE_OF_INCIDENT column
yr2010_present_df['DATE_OF_INCIDENT'] = yr2010_present_df['DATE_OF_INCIDENT'].dt.date()
```
